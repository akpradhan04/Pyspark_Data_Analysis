### Split and Indexing

#### SPLIT

```python
from pyspark.sql.functions import split
```

```python
df.withColumn('Outlet_Type', split('Outlet_Type', ' ')).limit(10).display()
```

#### INDEXING

```python
df.withColumn('Outlet_Type', split('Outlet_Type', ' ')[1]).limit(10).display()
```

#### EXPLODE

```python
from pyspark.sql.functions import explode
```

```python
df_exp = df.withColumn('Outlet_Type', split('Outlet_Type', ' '))
df_exploded = df_exp.withColumn('Outlet_Type', explode('Outlet_Type'))

df_exploded.count()
```

```python
from pyspark.sql.functions import array_contains
```

```python
df_exp.withColumn('Type1_flag',array_contains('Outlet_Type','Type1')).limit(10).display()
```
