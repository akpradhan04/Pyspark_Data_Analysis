### Schema Definition

```python
df.printSchema()
```

### Update Schema Defintion

```python
my_ddl_schema = '''
                    Item_Identifier STRING,
                    Item_Weight STRING,
                    Item_Fat_Content STRING, 
                    Item_Visibility DOUBLE,
                    Item_Type STRING,
                    Item_MRP DOUBLE,
                    Outlet_Identifier STRING,
                    Outlet_Establishment_Year INT,
                    Outlet_Size STRING,
                    Outlet_Location_Type STRING, 
                    Outlet_Type STRING,
                    Item_Outlet_Sales DOUBLE 

                ''' 
```

### Reload with modified Schema

```python
df = spark.read.format('csv').option('header', True)\
                        .schema(my_ddl_schema)\
                        .load('/Workspace/Users/email_address@gmail.com/Pyspark_Data_Analysis/datasets/BigMart Sales.csv')
```

```python
df.display()
```

### Changed Schema Definition

```python
df.printSchema()
```
