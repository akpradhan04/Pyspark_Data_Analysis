### Read CSV

```python
df = spark.read.format('csv').option('inferSchema', True)\
                        .option('header', True)\
                        .load('/Workspace/Users/email_address@gmail.com/Pyspark_Data_Analysis/datasets/BigMart Sales.csv')
```

### SELECT statements

Conventional Method

```python
df.select('Item_Identifier', 'Item_Weight', 'Item_Fat_Content').display()
```

Alternatively

```python
from pyspark.sql.functions import *
```

```python
# col function needs to be import
df.select(col('Item_Identifier'),col('Item_Weight'),col('Item_Fat_Content')).display()
```

### ALIAS

```python
df.select(col('Item_Identifier').alias('Item')).show(10)
```

### FILTER

Scenario 1 : Item Fat Content = Regular

```python
df.filter(df.Item_Fat_Content == 'Regular').display()

df.filter(df['Item_Fat_Content'] == 'Regular').show()
```

```python
df.filter(col("Item_Fat_Content") == "Regular").display()
```

Scenario 2: Item_Type = Soft Drinks and Item_Weight < 10

```python
df.filter((col('Item_Type') == "Soft Drinks") & (col('Item_Weight') < 10)).display()
```

Scenario 2: Outlet Size is Null and Outlet Location Type is Tier 1 or Tier 2

```python
df.filter((df.Outlet_Size).isNull() & (df.Outlet_Location_Type).isin(['Tier 1', 'Tier 2'])).display()
```

```python
df.filter((col('Outlet_Size').isNull()) & (col('Outlet_Location_Type').isin(['Tier 1', 'Tier 2']))).display()
```
