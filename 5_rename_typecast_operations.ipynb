```python
df = spark.read.format('csv').option('inferSchema', True)\
                        .option('header', True)\
                        .load('/Workspace/Users/email_address@gmail.com/Pyspark_Data_Analysis/datasets/BigMart Sales.csv')
```

### Rename Column

```python
df.withColumnRenamed('Item_Weight', 'Weight').display()
```

### Modify Columns

New Column (Constant)

```python
from pyspark.sql.functions import lit

# lit -> literal function , used to assign constants across the rows
df = df.withColumn('flag', lit('new'))
```

```python
df.limit(10).display()
```

New Column (Calculation)

```python
from pyspark.sql.functions import col


df.withColumn('multiply', col("Item_Weight") * col("Item_MRP")).limit(10).display()

# This is also acceptable approach 
# df.withColumn('multiply', df['Item_MRP'] * df['Item_Weight']).limit(10).display()
```

Modify existing column

```python
from pyspark.sql.functions import regexp_replace

df.withColumn('Item_Fat_content', regexp_replace(col('Item_Fat_Content'), 'Low Fat', 'LF'))\
    .withColumn('Item_Fat_content', regexp_replace(col('Item_Fat_Content'), 'Regular', 'Reg'))\
    .limit(10).display()
```

### Type Casting

```python
from pyspark.sql.functions import col, StringType

df.withColumn('Item_Weight', col('Item_Weight').cast(StringType())).limit(10).display()
```
