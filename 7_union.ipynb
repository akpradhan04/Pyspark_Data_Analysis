### Prepare DataFrame

```python
data1 = [(1, 'john'), (2, 'sam')]
schema1 = 'id STRING, name STRING'

df1 = spark.createDataFrame(data1, schema1)

data2 = [(3, 'jane'), (4, 'mary')]
schema2 = 'id STRING, name STRING'

df2 = spark.createDataFrame(data2, schema1)

```

```python
display(df1)
```

```python
display(df2)
```

### UNION

```python
df_union = df1.union(df2)
```

```python
df_union.display()
```
