```python
data1 = [(1, 'john'), (2, 'sam')]

df = spark.createDataFrame(data1, ['id', 'name'])
df.show()
```

### Date Functions

Current Date

```python
from pyspark.sql.functions import current_date, date_add, date_sub, datediff, date_format
```

```python

df = df.withColumn('current_date', current_date())

df.show()
```

Date Add and Sub

```python
df = df.withColumn('week_after', date_add('current_date', 7))
df.show()
```

```python
df = df.withColumn('week_before', date_sub('current_date', 7))

# df = df.withColumn('week_before', date_add('current_date', -7))

df.show()
```

Date Difference

```python
df = df.withColumn('date_diff', datediff('week_after', 'week_before'))
df.show()
```

Date Format

```python
df = df.withColumn('week_before', date_format('week_before', 'dd-MM-yyyy'))
df.show()
```
